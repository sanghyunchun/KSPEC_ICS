```python
import os
import glob
from gfa_getcrval import get_crvals_from_images

# 노트북에서는 __file__ 대신 현재 작업 디렉토리 사용
base_dir = os.getcwd()

test_dir = os.path.join(base_dir, "img", "astroimg")
images = sorted(glob.glob(os.path.join(test_dir, "*.fits")))

if not images:
    raise FileNotFoundError(f"No .fits files found in: {test_dir}")

images = images[:4]  # 4개만 테스트

print(f"[INFO] Found {len(images)} FITS files:")
for p in images:
    print(" -", p)

crval1_list, crval2_list = get_crvals_from_images(images, max_workers=4)

print("\n[RESULT]")
for p, cr1, cr2 in zip(images, crval1_list, crval2_list):
    print(f"{os.path.basename(p)}  CRVAL1={cr1}  CRVAL2={cr2}")
```
