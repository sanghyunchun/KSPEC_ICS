```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
cam=cv2.VideoCapture(0)
cam.set(cv2.CAP_PROP_FRAME_WIDTH, 2544)
cam.set(cv2.CAP_PROP_FRAME_HEIGHT, 1944)
```

```python
ret,frame=cam.read()
frame =  frame[:,:,::-1]

r_data = np.array(frame[:,:,0])
r_data=r_data[::-1]
g_data = np.array(frame[:,:,1])
g_data=g_data[::-1]
b_data = np.array(frame[:,:,2])
b_data=b_data[::-1]
        
img_data=g_data

plt.imshow(img_data,origin='lower')
```

```python
cam.release()
```
